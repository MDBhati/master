Data Cleaning: Clean the dataset to remove any duplicates, inconsistencies or irrelevant columns. For example, we can remove the "show_id" and "description" columns as they may not be relevant for our analysis.

```python
netflix = df.drop(['show_id','description'],axis=1)
```

```python
netflix = netflix.drop_duplicates()
```

```python
# understand variables
variables = pd.DataFrame(columns=['variable','Number of unique values','Values'])

for i,var in enumerate(netflix.columns):
  variables.loc[i] = [var,netflix[var].nunique(),netflix[var].unique()]

variables
```

Q.1 Identify missing values: Check if there are any missing values in the dataset and decide how to handle them. For example, if the "director" column has missing values, we can either delete those rows or fill in the missing values with "unknown".

```python
netflix.info()
```

```python
netflix.isna().sum()  # #  Director, cast, country, date_added, rating and duration Columns have null values.
```

```python
netflix['director']  = netflix['director'].fillna('unknown')
netflix['cast']  = netflix['cast'].fillna('unknown')
netflix['country']  = netflix['country'].fillna('unknown')
netflix['date_added']  = netflix['date_added'].fillna(netflix['release_year'],axis=0)
netflix['rating'] = netflix['rating'].fillna('unknown')
```

```python
netflix.isna().sum()
```

```python
netflix['duration'] = netflix['duration'].str.replace(' min','')
netflix['duration'] = netflix['duration'].str.replace(' seasons','')
```

```python
netflix.shape
```

Q.3 Data Exploration: Use Matplotlib to create visualizations that explore the data in meaningful ways. For example, we can create a bar chart to show the number of movies vs TV shows available on Netflix, or a line chart to show the trend in the number of new titles added each year.

```python
x = netflix['type'].unique()
y = netflix['type'].value_counts().values
```

```python
# 3. Data Exploration
# Bar plot showing numbers of TV shows vs movies
plt.bar(x,y)
plt.xlabel('Type',fontsize=1)
plt.ylabel('Count',fontsize=12)
plt.title('Count of Movies and TV Shows')
plt.show()
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
netflix['year_added'] = netflix['date_added'].dt.year
```

```python
movies = netflix[netflix['type']=='Movie']
tv_shows = netflix[netflix['type']=='TV Show']
```

```python
movies_added_year  = movies.year_added.value_counts()
tv_show_added_year = tv_shows.year_added.value_counts()
```

```python
plt.plot(movies_added_year.index,movies_added_year.values)
plt.xlabel("Year",fontsize=12)
plt.ylabel('Count')
plt.title("Number of movies added on Netflix")
plt.show()
```

```python
plt.plot(tv_show_added_year.index,tv_show_added_year.values,marker='o')
plt.xlabel("Year",fontsize=12)
plt.ylabel('Count')
plt.title("Number of TV Shows added on Netflix")
plt.show()
plt.show()
```

Q.4 Analyze Genres: Analyze the genres of movies and TV shows available on Netflix using bar charts or pie charts. For example, we can create a bar chart to show the top 10 most popular genres on Netflix.

```python
top10_movie_genre = movies.listed_in.value_counts().head(10)
top10_movie_genre_sort =top10_movie_genre.sort_values()
```

```python
top10_tv_shows_genre = tv_shows.listed_in.value_counts().head(10)
top10_tv_shows_genre_sort = top10_tv_shows_genre.sort_values()
```

```python
plt.figure(figsize=(15,8))
plt.barh(top10_tv_shows_genre_sort.index,top10_tv_shows_genre_sort.values)
plt.xlabel('Number of TV Shows',fontsize=15)
plt.ylabel('genre',fontsize=15)
plt.title('Number of TV Shows by genre',fontsize=22)
plt.show()
```

Q.5 Analyze Ratings: Analyze the ratings of movies and TV shows available on Netflix using histograms or box plots. For example, we can create a histogram to show the distribution of ratings on Netflix.

```python
rating = netflix.groupby('rating').agg(rating_count = ('rating','count')).sort_values('rating_count',ascending=False)
rating
```

```python
plt.figure(figsize=(14,7),dpi=150)
graph = sns.countplot(data=netflix,x='rating',order=rating.index)
plt.xticks(rotation=90)

for p in graph.patches:
  height = p.get_height()
  graph.text(p.get_x() + p.get_width()/2, height+20,height,ha='center',fontsize=8)

text = 'Top 3 most used rating :\n\n   1. TV-MA\n   2. TV-14\n   3. TV-PG'
plt.text(5,2500,text,horizontalalignment='left',fontsize=14,fontstyle='italic')
plt.title('Rating across Netflix',weight='semibold',fontsize=22)
plt.show()
```

Q.2 Fill in missing values: If there are missing values in the Age column, use appropriate techniques such as imputation or interpolation to fill in the missing values.

```python
class_age_dict  = dict(titanic.groupby('pclass')['age'].mean().round())
class_age_dict  # class 1 mean age 38 year ,class2 mean age 30 and class3 passenger mean age is 25
```

```python
titanic[titanic['age'].isna()].value_counts('pclass')  # class wise missing values in age column
```

```python
titanic['age']  = titanic['age'].fillna(titanic['pclass'].map(class_age_dict))
# fill age column missing value with respect to their passenger class mean age
```

```python
def age_category(age):
  if age<=10:
    category = '0-10'
  elif age <20:
    category ='11-20'
  elif age <=40:
    category = '21-40'
  elif age <=60:
    category ='41-60'
  else:
    category ='>60'
  return category

titanic['age_category'] = titanic.apply(lambda x: age_category(x['age']),axis=1)
titanic.head()
```
